```python
from tensorflow import keras

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras import layers, models, optimizers, regularizers
from keras.optimizers import Adam
```

```python
class MeanTeacherModel(models.Model):
    def __init__(self):
        super(MeanTeacherModel, self).__init__()
        self.teacher = models.Sequential([
            layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
            layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
            layers.Dense(3, activation='softmax')
        ])
        self.student = models.Sequential([
            layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
            layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
            layers.Dense(3, activation='softmax')
        ])

    def call(self, x, training=True):
        teacher_pred = self.teacher(x)
        student_pred = self.student(x)
        if training:
            return student_pred, teacher_pred
        else:
            return student_pred, student_pred, teacher_pred

def mean_teacher_loss(y_true, y_pred, y_teacher=None, alpha=0.5):
    print("y_true shape: ", y_true.shape)
    print("y_pred shape: ", y_pred.shape)
    
    supervised_loss = keras.losses.categorical_crossentropy(y_true, y_pred)
    
    if y_teacher is not None:
        unsupervised_loss = keras.losses.categorical_crossentropy(y_teacher, y_pred)
        return alpha * supervised_loss + (1 - alpha) * unsupervised_loss
    else:
        return supervised_loss


```

```python
from keras.optimizers import Adam
optimizer = keras.optimizers.Adam(learning_rate=0.001)
model = MeanTeacherModel()
model.compile(optimizer=optimizer, loss=mean_teacher_loss)
```

```python
# define the model
model = MeanTeacherModel()

# compile the model with the mean teacher loss function
model.compile(optimizer='adam', loss=mean_teacher_loss)

# test the model
model.fit(X_test_encoded, y_test_array, epochs=10, batch_size=32)

# generate predictions on the test set
y_pred = model.predict(X_test_encoded)


# Example: Convert y_pred probabilities to predicted class labels
predicted_labels = np.argmax(y_pred, axis=1)


# extract the NumPy array from the tuple
y_pred_array = y_pred[0]
# Print the predicted labels
print(y_pred_array.shape)


# Get the predicted class labels
labels = ['Depression', 'Anxiety', 'Insomnia']
y_pred_labels = [labels[i] for i in y_pred_array.argmax(axis=1)]


# Convert predicted labels to one-hot encoded format
predicted_labels_encoded = np.eye(len(labels))[np.argmax(y_pred_array, axis=1)]
print(predicted_labels_encoded.shape)



# Get the true labels
#true_labels = np.argmax(y_test_array, axis=1)

# Print the true labels
#print("True labels:", true_labels)
# Convert true labels to one-hot encoded format

y_true = y_test_array
print(y_true.shape)


# Create dictionaries to map labels to student names
depression_dict = {}
anxiety_dict = {}
insomnia_dict = {}

for i in range(len(X_test)):
    student_name = X_test.index[i]
    predicted_label = y_pred_labels[i]
    if predicted_label == 'Depression':
        depression_dict[student_name] = predicted_label
    elif predicted_label == 'Anxiety':
        anxiety_dict[student_name] = predicted_label
    elif predicted_label == 'Insomnia':
        insomnia_dict[student_name] = predicted_label


# Display the student names for each label
print('\n')
print('Depression students:', list(depression_dict.keys()))
print('\n')
print('Anxiety students:', list(anxiety_dict.keys()))
print('\n')
print('Insomnia students:', list(insomnia_dict.keys()))

# Count the total number of students in each catagory
print('\n')
num_depression_students = len(depression_dict)
print("Number of depression students:", num_depression_students)

num_anxiety_students = len(anxiety_dict)
print("Number of anxiety students:", num_anxiety_students)

num_insomnia_students = len(insomnia_dict)
print("Number of insomnia students:", num_insomnia_students)
```
